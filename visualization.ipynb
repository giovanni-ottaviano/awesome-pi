# Visualization

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
mpl.rcParams['axes.titlesize'] = 16
mpl.rcParams['axes.labelsize'] = 14
mpl.rcParams['legend.fontsize'] = 12
mpl.rcParams['xtick.labelsize'] = 12
mpl.rcParams['ytick.labelsize'] = 12
```

### 1. Monte Carlo: Circle method

A simple visualization of the results given by *Circle method*.

```python
blocks, pi_mean_circle, pi_stderr_circle = np.loadtxt('results/pi_circle.dat', dtype=float, unpack='true')
throws_per_block = 100000
N = blocks * throws_per_block

plt.errorbar(N, pi_mean_circle, yerr=pi_stderr_circle, label='MC result')
plt.hlines(y=np.pi, xmin=0, xmax=N[-1], linestyles='dashed', color='red', label='NumPy pi')
plt.ticklabel_format(axis="x", style="sci", scilimits=(0,0))
plt.title(f'Estimation of $\pi$ - Circle method', size=16)
plt.xlabel(f'MC steps (points)')
plt.ylabel(f'Estimation of $\pi$')
plt.legend(loc='lower right', fontsize=12)
plt.grid(True)
plt.show()
```

### 2. Monte Carlo: Buffon's needle method

A simple visualization of the results given by *Buffon's method*.

```python
blocks, pi_mean_buffon, pi_stderr_buffon = np.loadtxt('results/pi_buffon.dat', dtype=float, unpack='true')
throws_per_block = 100000
N = blocks * throws_per_block

plt.errorbar(N, pi_mean_buffon, yerr=pi_stderr_buffon, label='MC result')
plt.hlines(y=np.pi, xmin=0, xmax=N[-1], linestyles='dashed', color='red', label='NumPy pi')
plt.ticklabel_format(axis="x", style="sci", scilimits=(0,0))
plt.title(f'Estimation of $\pi$ - Buffon\'s method', size=16)
plt.xlabel(f'MC steps (needle throws)')
plt.ylabel(f'Estimation of $\pi$')
plt.legend(loc='lower right', fontsize=12)
plt.grid(True)
plt.show()
```

### 3. Function integration

A simple visualization of the results given by *Function integration method*.

```python
blocks, pi_mean_integral, pi_stderr_integral = np.loadtxt('results/pi_integral.dat', dtype=float, unpack='true')
throws_per_block = 100000
N = blocks * throws_per_block

plt.errorbar(N, pi_mean_integral, yerr=pi_stderr_integral, label='MC result')
plt.hlines(y=np.pi, xmin=0, xmax=N[-1], linestyles='dashed', color='red', label='NumPy pi')
plt.ticklabel_format(axis="x", style="sci", scilimits=(0,0))
plt.title(f'Estimation of $\pi$ - Integration', size=16)
plt.xlabel(f'MC steps')
plt.ylabel(f'Estimation of $\pi$')
plt.legend(loc='upper right', fontsize=12)
plt.grid(True)
plt.show()
```

### 4. Leibniz series with Shanks transformation

A simple visualization of the results given by the *transformed Leibniz series*.

```python
with open('results/pi_weps.dat', 'r') as f:
    pi_leibniz = float(f.read())

print("Result:", pi_leibniz)
print(f"Percentual difference: {np.abs(pi_leibniz - np.pi)/np.pi}")
```

## Result comparison

It is useful to show a representation of the results for all methods compared

```python
plt.errorbar(1, pi_mean_circle[-1], yerr=pi_stderr_circle[-1], label='Circle')
plt.errorbar(2, pi_mean_buffon[-1], yerr=pi_stderr_buffon[-1], label='Buffon')
plt.errorbar(3, pi_mean_integral[-1], yerr=pi_stderr_integral[-1], label='Integral')
plt.scatter(4, pi_leibniz, label='Leibniz')
plt.hlines(y=np.pi, xmin=1, xmax=4, linestyles='dashed', color='red', label='NumPy pi')
plt.title(f'Estimation comparison', size=16)
plt.xlabel(f'Method')
plt.ylabel(f'Estimation of $\pi$')
# plt.legend(loc='upper right', fontsize=10)
plt.xticks(range(1,5), labels=['cicle','buffon','integral','leibniz'], rotation=45)
plt.show()
```
